```python
X=data.iloc[:,:-1]
Y=data['Heart Disease']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

###  logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
x_pred=LR.predict(x_train)
print(f"accuracy on training data {accuracy_score(y_train,x_pred)}")
print(classification_report(y_train,x_pred))
```

```python
y_pred=LR.predict(x_test)
print(f"accuracy on test data {accuracy_score(y_pred,y_test)}")
print(classification_report(y_test,y_pred))
```

### svc

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
x_predict=sv.predict(x_train)
print(f"accuracy on test data {accuracy_score(x_predict,y_train)}")
print(classification_report(y_train,x_predict))
```

```python
y_predict=sv.predict(x_test)
print(f"accuracy score of test data {accuracy_score(y_predict,y_test)}")
print(classification_report(y_predict,y_test))
```

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(x_train, y_train)
prediction=clf.predict(x_train)
print(f"accuracy score of training data {accuracy_score(y_train,prediction)}")
print(classification_report(y_train,prediction))
```

```python
y_predict=clf.predict(x_test)
print(f"accuracy score of test data is {accuracy_score(y_predict,y_test)}")
print(classification_report(y_predict,y_test))
```

### Naive bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
x_predict=gnb.predict(x_train)
print(f"accuracy score of training data {accuracy_score(y_train,x_predict)}")
print(classification_report(y_train,x_predict))
```

```python
y_predict=gnb.predict(x_test)
print(f"accuracy score of test data {accuracy_score(y_test,y_predict)}")
print(classification_report(y_test,y_predict))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
ld=LinearDiscriminantAnalysis()
ld.fit(x_train,y_train)
x_predict=ld.predict(x_train)
print(f"accuracy score of training data {accuracy_score(y_train,x_predict)}")
```

```python
y_predict=ld.predict(x_test)
```
